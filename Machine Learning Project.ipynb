```python
#Importing Anime Data

animes_df = pd.read_csv("anime.csv")
animes_df
```

```python
#dropping columns that won't be used in model

animes_skimmed = animes_df.drop(['Japanese name',
       'Type', 'Episodes', 'Aired', 'Premiered', 'Producers', 'Licensors',
       'Studios', 'Source', 'Duration', 'Rating', 'Ranked', 'Popularity',
       'Members', 'Favorites', 'Watching', 'Completed', 'On-Hold', 'Dropped',
       'Plan to Watch', 'Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6',
       'Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1','Name', 'Score', 'English name'], axis = 1)

animes_skimmed
```

```python
#Reshaping table such that, each unique value of "Genres" column is made it's own column with binary value representing 
#Whether a specific anime is that genre or not.

genres_split_out = animes_skimmed['Genres'].str.get_dummies(', ')
genres_split_out = pd.concat([animes_skimmed, genres_split_out], axis=1)
genres_split_out = genres_split_out.drop('Genres', axis = 1)
genres_split_out
```

```python
# For the purpose of testing, pull all the animes seen by a user in the user table with user_ID

def generate_single_user_ratings(user_id):
    user_ratings = ratings_df[ratings_df['user_id'] == user_id]['rating'].reset_index()
    user_ratings.rename(columns={'anime_id': 'rating'})
    return user_ratings
```

```python
user_ratings = generate_single_user_ratings(1)
user_ratings
```

```python
#Performs a Left Join on user ratings and genres split out, to get the binary value for every genre that applies to 
#all the anime the user has seen

def generate_merged_df(genres_split_out, user_ratings):
    df1_filtered = genres_split_out[genres_split_out['MAL_ID'].isin(user_ratings['index'])]
    merged_df = pd.merge(df1_filtered, user_ratings, left_on='MAL_ID', right_on='index')
    merged_df.drop('index', axis=1, inplace=True)
    return merged_df
    
```

```python
merged_df = generate_merged_df(genres_split_out, user_ratings)
merged_df
```

```python
from sklearn.model_selection import train_test_split

#This function will take our training data and split it up for training

def split_user_data_train_test(merged_df):
    X = merged_df.drop(columns=["MAL_ID", "rating"])
    y = merged_df["rating"]
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
    return x_train, x_test, y_train, y_test
    
```

```python
#This function by using animes_df and anime_ratings will take an anime name and return the most popular anime 
#For each genre that anime is part of

def find_similar_anime(anime_id):
    anime_genres = animes_df[animes_df['MAL_ID'] == anime_id]['Genres'].iloc[0].split(", ")
    anime_ids = []
    for i in range(len(anime_genres)):
        genre = anime_genres[i]
        similar_anime = animes_df[animes_df['Genres'].str.contains(genre)]
        anime_ids.extend(list(similar_anime['MAL_ID']))
    anime_ids = list(set(anime_ids))
    anime_ids.remove(anime_id)
    anime_ratings = ratings_df[ratings_df['anime_id'].isin(anime_ids)]
    anime_ratings_count = anime_ratings.groupby('anime_id')['rating'].count()
    anime_with_most_ratings = anime_ratings_count.nlargest(5)
    anime_info = animes_df[animes_df['MAL_ID'].isin(anime_with_most_ratings.index)][['Name', 'MAL_ID', 'Genres']]
    return anime_info
```

```python
from sklearn.naive_bayes import BernoulliNB

# Perform Naive Bayes with Bernoulli distribution, takes in training and testing data and returns summed error, and 
# and label predicion Accuracy

#This naive bayes uses the probability that a user will like an anime with any given genre as priors. And 
#Uses the users ratings of 1-10 as labels. For testing data NB will look at any anime and try and predict 
#the rating that a user will give that anime. The "error_total" is the sum of absolute differences between the predicted
#rating and the actual. "correct_counter/iterations" is the accuracy, or what percentage of predictions are dead on.
#Choose to include both metrics to see if bernoulli would perform better than gaussian. 


def NB_Bernoulli(x_train, y_train, x_test, y_test):
    model = BernoulliNB()
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    error_total = 0
    correct_counter = 0
    iterations = 0
    for x in range(len(y_pred)-1):
        iterations += 1
        if (list(y_pred)[x] != list(y_test)[x]):
            error_total += abs(list(y_pred)[x] - list(y_test)[x])
        else:
            correct_counter += 1
    return error_total, correct_counter/iterations 
    
```

```python
NB_Bernoulli(x_train, y_train, x_test, y_test)
#The first value is the total absolute value of the error, The second is how often it was exactly correct with label
```

```python
from sklearn.naive_bayes import GaussianNB


#Same setup as Bernoulli, in terms of arguemnts and error metrics. 


def NB_Gaussian(x_train, y_train, x_test, y_test):
    gnb = GaussianNB()
    gnb.fit(x_train,y_train)
    y_pred = gnb.predict(x_test)
    error_total = 0
    correct_counter = 0
    iterations = 0
    for x in range(len(y_pred)-1):
        iterations += 1
        if (list(y_pred)[x] != list(y_test)[x]):
            error_total += abs(list(y_pred)[x] - list(y_test)[x])
        else: 
            correct_counter += 1 
            
    return error_total, correct_counter/iterations 
```

```python
#This function will perform Gaussian NB on multiple user and average the total error and accuracy to give us better idea
#of how model is working. 

def find_average_error_for_n_users_Gaussian(number_of_users):
    total_error = 0
    accuracy_error = 0
    iterations = 0
    for x in range(number_of_users):
        user_ratings = generate_single_user_ratings(x)
        merged_df = generate_merged_df(genres_split_out, user_ratings)
        if(merged_df.empty == True):
            continue
        else:
            iterations += 1
            x_train, x_test, y_train, y_test = split_user_data_train_test(merged_df)
            temp_error, temp_accuracy_error = NB_Gaussian(x_train, y_train, x_test, y_test)
            total_error += temp_error
            accuracy_error += temp_accuracy_error
        
    return total_error/iterations, accuracy_error/iterations
        
```

```python
#This function will perform Bernoulli NB on multiple user and average the total error and accuracy to give us better idea
#of how model is working. 

def find_average_error_for_n_users_Bernoulli(number_of_users):
    total_error = 0
    accuracy_error = 0
    iterations = 0
    for x in range(number_of_users):
        iterations += 1
        user_ratings = generate_single_user_ratings(x)
        merged_df = generate_merged_df(genres_split_out, user_ratings)
        if(merged_df.empty == True):
            continue
        else: 
            iterations += 1
            x_train, x_test, y_train, y_test = split_user_data_train_test(merged_df)
            temp_error, temp_accuracy_error = NB_Bernoulli(x_train, y_train, x_test, y_test)
            total_error += temp_error
            accuracy_error += temp_accuracy_error
        
    return total_error/iterations, accuracy_error/iterations 
```

Since NB doesn't have a hyperparam to tune, We instead chose to try two different distributions to generate priors. Bernoulli and Guassaun. 

While looking at the results for multiple users, It looks like Guassian while having better accuracy than Bernoulli, has less total error. Eventhough Guassian made the exact correct prediction more often than Bernoulli, Bernoulli has way less total error so is closer in general to the correct ratings eventhough it's not 100% on target as often as Gausssian.

Therefore for our recommender system we will implement Bernoulli!

```python
#Helper function to generate df form for user

def get_ratings_dataframe(animeNames, ratings, animes_df):
    anime_id_list = []
    for name in animeNames:
        anime_id = animes_df.loc[animes_df['Name'] == name, 'MAL_ID'].iloc[0]
        anime_id_list.append(anime_id)

    ratings_dict = {'index': anime_id_list, 'rating': ratings}
    ratings_df = pd.DataFrame(ratings_dict)

    return ratings_df
    
    
    
```

```python
#helper function to get anime_df with all the animes except the ones a user has scene to predict ratings for

def filter_animes(user_ratings, animes_df):
    filtered_animes = animes_df[animes_df['MAL_ID'].isin(user_ratings) == False]
    anime_ids = filtered_animes['MAL_ID']
    filtered_animes = filtered_animes.drop('MAL_ID', axis = 1)
    return filtered_animes, anime_ids
```

```python
#This is the reccommender! takes in the user data in the form of x_train, y_train, and 
#the number of recommdations you want to recieve

#This function will predict the rating for every anime the user hasn't seen, and get the num_of_reccommendations
#from sorted list predicted animes ranked desending with predicted highest rated anime at the top 

def NB_Bernoulli_Recommender(x_train, y_train, num_of_recommendations):
    model = BernoulliNB()
    model.fit(x_train, y_train)
    
    prediction_df, anime_ids = filter_animes(x_train, genres_split_out)
    
    y_pred = model.predict(prediction_df)

    recommender_dict = {}
    
    for x in range(len(y_pred)-1):
        recommender_dict[anime_ids[x]] = y_pred[x]
        
        
    items = recommender_dict.items()
    sorted_items = sorted(items, key=lambda x: x[1], reverse = True)
    
    recommendations = []*num_of_recommendations
    
    for x in range(num_of_recommendations):
        print(str(animes_df[animes_df['MAL_ID'] == sorted_items[x][0]]['Name'].iloc[0]))
              
    
```
